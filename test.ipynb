```python
loan_amount=100000
year_rate=3.5
year_priod=7
```

```python
month_pay=int(loan_amount/(year_priod*12))
month_pay
```

## 計算每月應還的本金+利息包含本利和跟餘額

```python
for i in range(year_priod*12):
    print(f'期別:{i+1}')
```

```python
interest=round(loan_amount*(year_rate/100/12))
interest
```

```python
amount=loan_amount
months=year_priod*12
month_rate=year_rate/100/12
#固定還款金額
month_pay=int(loan_amount/months)
#期數
for i in range(months):
    #計算每月利息   
    interest=round(amount*month_rate)
    amount-=month_pay
    #最後一期
    if i==months-1:
        print(f'期別:{i+1} 本金:{month_pay+amount} 利息:{interest} 本利和:{month_pay+interest+amount} 餘額:{0}') 
    else:
        print(f'期別:{i+1} 本金:{month_pay} 利息:{interest} 本利和:{month_pay+interest} 餘額:{amount}') 
    
```
